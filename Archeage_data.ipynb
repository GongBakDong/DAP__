### **거리 계산기**

1.   각 지역의 기준의 판매 지역은 교역소/ 제작 지역은 주민회관 이다.
2.   거리는 직선거리로 계산한다.
3.   해당 좌표 값은 아키에이지 홈페이지 https://archeage.xlgames.com/wikis/%EC%A2%8C%ED%91%9C 에서 제공하는 육분의 좌표 값을 실수 값으로 변환한 것이다.




```python
import math
from prettytable import PrettyTable

points = {
    '솔즈리드 반도': (-14.4356, -6.0261),
    '두 왕관': (-17.5089, -7.8414),
    '십자별 평원': (-16.73138, -5.97972),
    '가랑돌 평원': (-15.00777, -9.14166),
    '그위오니드 숲': (-12.83194, -10.40138),
    '릴리엇 구릉지': (-12.89111, -8.67361),
    '마리아노플': (-16.39361, -10.57944),
    '청동 바위산': (-14.90805, -14.91805),
    '태초의요람': (-15.72638, -13.61055),
    '하늬마루': (-18.755, -14.855),
    '하얀 숲':(-12.12916, -11.13083),
    '긴 모래톱': (-18.62055, -10.51833),
    '지옥 늪지대': (-18.39888, -13.68972),
    '뼈의 땅':(-11.05138, -10.76972)

}

header = ['Name', '솔즈리드 반도', '두 왕관', '십자별 평원']

solzreed_dist = []
crown_dist = []
star_dist = []

for name, coord in points.items():
    dist_solzreed = round(math.sqrt((points['솔즈리드 반도'][0]-coord[0])**2 + (points['솔즈리드 반도'][1]-coord[1])**2), 5)
    solzreed_dist.append(dist_solzreed)
    dist_crown = round(math.sqrt((points['두 왕관'][0]-coord[0])**2 + (points['두 왕관'][1]-coord[1])**2), 5)
    crown_dist.append(dist_crown)
    dist_star = round(math.sqrt((points['십자별 평원'][0]-coord[0])**2 + (points['십자별 평원'][1]-coord[1])**2), 5)
    star_dist.append(dist_star)

data = list(zip(points.keys(), solzreed_dist, crown_dist, star_dist))

table = PrettyTable()
table.field_names = header
for row in data:
    table.add_row(row)



table.align = 'c'
table.align['Name'] = 'l'
table.junction_char = '|'
table.horizontal_char = '-'
table.header_style = 'upper'
table.title = '거리표'
table.title_style = 'upper'
table.padding_width = 1
print(table)
```

```python
product_A = int(input("현재 일반 특산품 재고 수: "))
product_B = int(input("현재 상단 특산품 재고 수: "))
product_C = int(input("현재 숙성 특산품 재고 수: "))

min_product_A = 50
min_product_B = 30
min_product_C = 10

price_increase = 0.05
price_limit = 1.3

product_met_count = 0

while product_A >= min_product_A and product_B >= min_product_B and product_C >= min_product_C:
    product_met_count += 1
    price_increase_total = price_increase * product_met_count
    new_price = 1 + price_increase_total
    if new_price <= price_limit:
        print(f"물가 상승 ({price_increase*100}% x {product_met_count})")
        if product_C < min_product_C:
            shortage = min_product_C - product_C
            if product_A >= min_product_A * 2:
                converted = shortage // 5
                converted = min(converted, product_A // (min_product_A * 2))
                product_A -= converted * min_product_A * 2
                product_C += converted * 5
            elif product_B >= min_product_B * 2:
                converted = shortage // 5
                converted = min(converted, product_B // (min_product_B * 2))
                product_B -= converted * min_product_B * 2
                product_C += converted * 5
            else:
                break
        else:
            product_A -= min_product_A * 2
            product_B -= min_product_B * 2
            product_C -= min_product_C
    else:
        print("물가 제한 130%")
        break

if product_A <= 0 and product_B <= 0 and product_C <= 0:
    print("물가 상승")

print(f"잔여 재고 수: A={product_A}, B={product_B}, C={product_C}")
```
